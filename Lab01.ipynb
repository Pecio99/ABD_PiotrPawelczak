# Zadanie 3

Zdefiniuj poniższą funkcję i sporządź jej wykres dla argumentów z danego przedziału:

$f(x)=x^2+5$

$x>-1$ oraz $x<1$ <br>
$x>-6$ oraz $x<6$ <br>
$x>0$ oraz $x<5$ <br>

Wspierając się dokumentacją Matplotlib. Dodaj do wykresu etykiety osi, tytuły wykresów i legendy.

Import potrzebnych bibliotek

```python
import matplotlib.pyplot as plt
import numpy as np
```

Implementacja funkcji

```python
def f(x):
    return x**2 + 5
```

Utworzenie tablic argumentów

```python
x_1 = np.linspace(-1, 1, 100)
x_2 = np.linspace(-6, 6, 100)
x_3 = np.linspace(0, 5, 100)
```

Utworzenie wykresów

```python
plt.plot(x_1, f(x_1), label='f(x)')
plt.grid()
plt.title('Wykres w zakresie [-1, 1]')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

```python
plt.plot(x_2, f(x_2))
plt.grid()
plt.title('Wykres w zakresie [-6, 6]')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

```python
plt.plot(x_3, f(x_3))
plt.grid()
plt.title('Wykres w zakresie [0, 5]')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```
